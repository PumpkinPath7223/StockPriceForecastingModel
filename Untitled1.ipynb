```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import root_mean_squared_error
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
ticker = 'AAPL'
df = yf.download(ticker, '2020-01-01')
```

```python
df
```

```python
df.Close.plot(figsize =(12,8))
```

```python
scaler = StandardScaler()

df['Close'] = scaler.fit_transform(df['Close'])
```

```python
df.Close
```

```python
seq_length = 30
data = []

for i in range(len(df) - seq_length):
    data.append(df["Close"].values[i : i + seq_length])
data = np.array(data)
```

```python
train_size = int(0.8 * len(data))

x_train = torch.tensor(data[:train_size, :-1, :]).type(torch.Tensor).to(device)
y_train = torch.tensor(data[:train_size, -1, :]).type(torch.Tensor).to(device)
x_test = torch.tensor(data[train_size:, -1, :]).type(torch.Tensor).to(device)
y_test = torch.tensor(data[train_size:, -1, :]).type(torch.Tensor).to(device)

# Reshape input for LSTM: (batch_size, sequence_length, input_size)
# x_train = x_train.unsqueeze(-1)
# x_test = x_test.unsqueeze(-1)
```

```python
class PredictionModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(PredictionModel, self).__init__()
        self.num_layers = num_layers
        self.hidden_dim = hidden_dim
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        
    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim, device=device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_dim, device=device)
        
        out, (hn, cn) = self.lstm(x, (h0.detach(), c0.detach()))
        out = self.fc(out[:, -1, :])
        return out
```

```python
model = PredictionModel(input_dim=1, hidden_dim=32, num_layers=2, output_dim=1).to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
num_epochs = 200

for i in range(num_epochs):
    y_train_pred = model(x_train)
    loss = criterion(y_train_pred, y_train)
    
    if i % 25 == 0:
        print(i,loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    
```

```python
model.eval()

# Reshape x_test to match the expected dimensions
# If x_test is a single sample, add batch dimension
if len(x_test.shape) == 2:
    x_test = x_test.unsqueeze(0)  # Add batch dimension
# If x_test has too many dimensions, reshape appropriately
elif len(x_test.shape) > 3:
    # Reshape based on your specific model requirements
    # Example: x_test = x_test.view(batch_size, seq_length, features)
    pass

# Make prediction with proper dimensions
with torch.no_grad():  # Disable gradient calculation for inference
    y_test_pred = model(x_test)
    
    
if isinstance(y_train_pred, torch.Tensor):
    y_train_pred = y_train_pred.detach().cpu().numpy()
    
# For y_test_pred
if isinstance(y_test_pred, torch.Tensor):
    y_test_pred = y_test_pred.detach().cpu().numpy()
    
# For y_train_true
if isinstance(y_train, torch.Tensor):
    y_train_true = y_train.detach().cpu().numpy()
    
# For y_test_true
if isinstance(y_test, torch.Tensor):
    y_test_true = y_test.detach().cpu().numpy()
    
    
y_train_pred_inv = scaler.inverse_transform(y_train_pred)
y_test_pred_inv = scaler.inverse_transform(y_test_pred)
y_train_true_inv = scaler.inverse_transform(y_train_true)
y_test_true_inv = scaler.inverse_transform(y_test_true)
```

```python
train_rmse = root_mean_squared_error(y_train[:, 0], y_train_pred[:, 0])
test_rmse = root_mean_squared_error(y_test[:, 0], y_test_pred[:, 0])
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
```

```python
fig = plt.figure(figsize=(12, 10))
gs = fig.add_gridspec(4, 1)

# Actual vs Predicted plot (top 3/4)
ax1 = fig.add_subplot(gs[:3, 0])

ax1.plot(df.iloc[-len(y_test):].index, y_test, label="Actual Price", color="blue")


ax1.plot(df.iloc[-len(y_test):], y_test, label="Predicted Price", color="green")
ax1.legend()
plt.title(f"{ticker} Stock Price Prediction")
plt.xlabel("Date")
plt.ylabel("Price")
ax1 = df.Close.plot(x_compat=True)
# Error plot (bottom 1/4)
ax2 = fig.add_subplot(gs[3, 0])
ax2.axhline(test_rmse, color="blue", linestyle="--", label="RMSE")
ax2.plot(df[-len(y_test):].index, abs(y_test - y_test), label="Prediction Error", color="red")
ax2.legend()
ax2 = df.Close.plot(x_compat=True)

plt.title("Prediction Error")
plt.xlabel("Date")
plt.ylabel("Error")


plt.tight_layout()
plt.show()
```
